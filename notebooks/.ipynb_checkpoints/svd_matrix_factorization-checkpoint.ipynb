````python
# ── Re-define evaluation functions (same as baseline notebook) ────────────────

def get_relevant_items(user_id, threshold=4.0):
    relevant = test_df[
        (test_df['user_id'] == user_id) &
        (test_df['rating'] >= threshold)
    ]['movie_id']
    return set(relevant)

def precision_at_k(recommended_ids, relevant_ids, k):
    if k == 0: return 0.0
    hits = len(set(list(recommended_ids)[:k]) & relevant_ids)
    return hits / k

def recall_at_k(recommended_ids, relevant_ids, k):
    if len(relevant_ids) == 0: return 0.0
    hits = len(set(list(recommended_ids)[:k]) & relevant_ids)
    return hits / len(relevant_ids)

def ndcg_at_k(recommended_ids, relevant_ids, k):
    if len(relevant_ids) == 0: return 0.0
    recommended_k = list(recommended_ids)[:k]
    dcg  = sum(1.0 / np.log2(i + 2) for i, item in enumerate(recommended_k) if item in relevant_ids)
    idcg = sum(1.0 / np.log2(i + 2) for i in range(min(k, len(relevant_ids))))
    return dcg / idcg if idcg > 0 else 0.0

def evaluate_model(recommend_fn, user_ids, k=10, sample_size=500, random_state=42):
    np.random.seed(random_state)
    sample_users = np.random.choice(user_ids, size=min(sample_size, len(user_ids)), replace=False)
    precisions, recalls, ndcgs = [], [], []
    for user_id in sample_users:
        relevant = get_relevant_items(user_id)
        if len(relevant) == 0:
            continue
        recs_df = recommend_fn(user_id, n=k)
        rec_ids = list(recs_df['movie_id'])
        precisions.append(precision_at_k(rec_ids, relevant, k))
        recalls.append(recall_at_k(rec_ids, relevant, k))
        ndcgs.append(ndcg_at_k(rec_ids, relevant, k))
    return {
        'precision@k': np.mean(precisions),
        'recall@k':    np.mean(recalls),
        'ndcg@k':      np.mean(ndcgs),
        'n_users':     len(precisions),
    }

def popularity_recommend(user_id, n=10):
    already_seen = set(train_df[train_df['user_id'] == user_id]['movie_id'])
    recs = (movie_stats[~movie_stats['movie_id'].isin(already_seen)]
            .head(n)[['movie_id', 'title', 'bayesian_avg']]
            .rename(columns={'bayesian_avg': 'score'}))
    recs['source'] = 'popularity'
    return recs.reset_index(drop=True)

print("✅ Evaluation functions ready")
````

## 2. SVD intuition — what it actually does

Before training, let's understand the core idea visually.

The rating matrix **R** (users × movies) is 95% empty. SVD factorises it into:

```
R  ≈  U  ×  V^T
```

- **U** — a (users × k) matrix. Each row = one user's taste in k hidden dimensions
- **V** — a (movies × k) matrix. Each row = one movie's profile in k hidden dimensions  
- **k** — number of latent factors (a hyperparameter we tune)

To predict user `u`'s rating for movie `i`:  
`r̂(u,i) = μ + b_u + b_i + U[u] · V[i]`

Where `μ` is the global mean, `b_u` is the user bias, `b_i` is the item bias.

**Nobody tells SVD what the k dimensions mean.** It discovers them from patterns. Factor 1 might end up representing "action vs drama preference." Factor 2 might represent "mainstream vs indie taste." We never know — and we don't need to.


````python
# ── Visualise the factorisation idea on a tiny toy example ──────────────────

np.random.seed(42)

# Toy 5×6 rating matrix (0 = missing)
toy_R = np.array([
    [5, 4, 0, 0, 1, 2],
    [4, 0, 0, 1, 2, 0],
    [0, 0, 4, 5, 0, 0],
    [0, 1, 5, 4, 0, 3],
    [1, 0, 4, 0, 5, 4],
])

fig, axes = plt.subplots(1, 3, figsize=(13, 3))

# Original matrix
im0 = axes[0].imshow(toy_R, cmap='Blues', vmin=0, vmax=5, aspect='auto')
axes[0].set_title('R — original (0 = missing)')
axes[0].set_xlabel('Movies →')
axes[0].set_ylabel('Users →')
plt.colorbar(im0, ax=axes[0], label='Rating')
for i in range(toy_R.shape[0]):
    for j in range(toy_R.shape[1]):
        val = toy_R[i, j]
        axes[0].text(j, i, str(val) if val > 0 else '?',
                     ha='center', va='center', fontsize=11,
                     color='white' if val > 3 else '#333')

# User matrix U (k=2 factors)
U = np.array([[0.9, 0.1], [0.8, 0.2], [0.1, 0.9], [0.2, 0.8], [0.3, 0.7]])
im1 = axes[1].imshow(U, cmap='Purples', vmin=0, vmax=1, aspect='auto')
axes[1].set_title('U — user factors (k=2)')
axes[1].set_xlabel('Latent factor →')
axes[1].set_ylabel('Users →')
axes[1].set_xticks([0, 1])
axes[1].set_xticklabels(['F1: action?', 'F2: drama?'], fontsize=9)
plt.colorbar(im1, ax=axes[1])
for i in range(U.shape[0]):
    for j in range(U.shape[1]):
        axes[1].text(j, i, f'{U[i,j]:.1f}',
                     ha='center', va='center', fontsize=10,
                     color='white' if U[i,j] > 0.6 else '#333')

# Movie matrix V
V = np.array([[0.9,0.1],[0.8,0.2],[0.1,0.9],[0.2,0.85],[0.85,0.15],[0.3,0.7]])
im2 = axes[2].imshow(V, cmap='Greens', vmin=0, vmax=1, aspect='auto')
axes[2].set_title('V — movie factors (k=2)')
axes[2].set_xlabel('Latent factor →')
axes[2].set_ylabel('Movies →')
axes[2].set_xticks([0, 1])
axes[2].set_xticklabels(['F1', 'F2'], fontsize=9)
plt.colorbar(im2, ax=axes[2])
for i in range(V.shape[0]):
    for j in range(V.shape[1]):
        axes[2].text(j, i, f'{V[i,j]:.2f}',
                     ha='center', va='center', fontsize=9,
                     color='white' if V[i,j] > 0.6 else '#333')

plt.tight_layout()
plt.savefig('../data/plot_svd_intuition.png', bbox_inches='tight')
plt.show()

print("📌 KEY INTUITION:")
print("  User 1 (row 0) has high F1, low F2 → action lover")
print("  User 3 (row 2) has low F1, high F2 → drama lover")
print("  Movie 1 (row 0 of V) has high F1, low F2 → action movie")
print()
print("  Predicted rating = U[user] · V[movie] (dot product)")
print(f"  User1 × Movie1 = {np.dot(U[0], V[0]):.2f}  → high (action lover × action movie ✅)")
print(f"  User1 × Movie3 = {np.dot(U[0], V[2]):.2f}  → low  (action lover × drama movie ✅)")
print(f"  User3 × Movie3 = {np.dot(U[2], V[2]):.2f}  → high (drama lover × drama movie ✅)")
````

## 3. Prepare data for Surprise

`scikit-surprise` has its own data format. We need to convert our pandas DataFrames into Surprise's `Dataset` object.


````python
from surprise import Dataset, Reader

# Surprise needs to know the rating scale
reader = Reader(rating_scale=(1, 5))

# Build Surprise datasets from our pre-split train/test DataFrames
# Surprise expects columns: user, item, rating
surprise_train_df = train_df[['user_id', 'movie_id', 'rating']]
surprise_test_df  = test_df[['user_id', 'movie_id', 'rating']]

# Load into Surprise format
full_dataset   = Dataset.load_from_df(surprise_train_df, reader)
trainset       = full_dataset.build_full_trainset()

# Build testset from our held-out test_df
# Surprise testset is a list of (uid, iid, rating) tuples
testset = [(str(row.user_id), str(row.movie_id), row.rating)
           for row in surprise_test_df.itertuples()]

print(f"Surprise trainset:")
print(f"  Users  : {trainset.n_users:,}")
print(f"  Items  : {trainset.n_items:,}")
print(f"  Ratings: {trainset.n_ratings:,}")
print()
print(f"Testset : {len(testset):,} ratings")
print()
print("✅ Data ready for Surprise")
````

## 4. Train SVD — default hyperparameters first

Always start with defaults to get a baseline score, then tune.


````python
# ── Train SVD with default hyperparameters ────────────────────────────────────
print("Training SVD (default params)...")
print("  n_factors=100, n_epochs=20, lr_all=0.005, reg_all=0.02")
print()

svd_default = SVD(n_factors=100, n_epochs=20, lr_all=0.005, reg_all=0.02,
                  random_state=42, verbose=False)

t0 = time.time()
svd_default.fit(trainset)
train_time = time.time() - t0

# Evaluate on test set
predictions = svd_default.test(testset)
rmse = accuracy.rmse(predictions, verbose=False)
mae  = accuracy.mae(predictions,  verbose=False)

print(f"Training time : {train_time:.1f}s")
print(f"RMSE          : {rmse:.4f}  (target: < 0.90)")
print(f"MAE           : {mae:.4f}")
print()
print("📌 RMSE = average error in predicted star rating.")
print("   RMSE of 0.87 means we're typically off by 0.87 stars.")
print("   But remember: RMSE alone doesn't tell us if RANKING is good.")
print("   We'll measure NDCG@10 later — that's what actually matters.")
````

## 5. Hyperparameter tuning with GridSearchCV

Three key hyperparameters to tune:
- `n_factors` — number of latent dimensions. More factors = more expressive but slower and more prone to overfitting
- `lr_all` — learning rate. How big each gradient step is
- `reg_all` — regularisation. Penalises large weights to prevent overfitting


````python
# ── Grid search over key hyperparameters ─────────────────────────────────────
# Note: GridSearchCV does k-fold CV internally — this will take ~5-10 minutes

print("Running GridSearchCV (3-fold CV)...")
print("This takes ~5-10 minutes on a laptop. Go make chai ☕")
print()

param_grid = {
    'n_factors': [50, 100, 150],
    'n_epochs':  [20, 30],
    'lr_all':    [0.005, 0.010],
    'reg_all':   [0.02, 0.1],
}

gs = GridSearchCV(
    SVD,
    param_grid,
    measures=['rmse', 'mae'],
    cv=3,
    n_jobs=-1,      # use all CPU cores
    refit=False,
    joblib_verbose=0,
)

full_data_for_cv = Dataset.load_from_df(
    train_df[['user_id', 'movie_id', 'rating']], reader
)

t0 = time.time()
gs.fit(full_data_for_cv)
elapsed = time.time() - t0

print(f"Grid search done in {elapsed:.0f}s")
print()
print(f"Best RMSE  : {gs.best_score['rmse']:.4f}")
print(f"Best params: {gs.best_params['rmse']}")
````

````python
# ── Visualise the grid search results ────────────────────────────────────────

results_df = pd.DataFrame(gs.cv_results)

# Plot: RMSE vs n_factors for each lr_all
fig, axes = plt.subplots(1, 2, figsize=(13, 4))

for lr in param_grid['lr_all']:
    subset = results_df[results_df['params'].apply(lambda p: p['lr_all'] == lr)]
    # Group by n_factors, take mean RMSE
    nf_rmse = subset.groupby(subset['params'].apply(lambda p: p['n_factors']))['mean_test_rmse'].min()
    axes[0].plot(nf_rmse.index, nf_rmse.values, marker='o', label=f'lr={lr}')

axes[0].set_title('RMSE vs n_factors (best reg per combo)')
axes[0].set_xlabel('n_factors')
axes[0].set_ylabel('Mean CV RMSE')
axes[0].legend()

for reg in param_grid['reg_all']:
    subset = results_df[results_df['params'].apply(lambda p: p['reg_all'] == reg)]
    nf_rmse = subset.groupby(subset['params'].apply(lambda p: p['n_factors']))['mean_test_rmse'].min()
    axes[1].plot(nf_rmse.index, nf_rmse.values, marker='o', label=f'reg={reg}')

axes[1].set_title('RMSE vs n_factors (best lr per combo)')
axes[1].set_xlabel('n_factors')
axes[1].set_ylabel('Mean CV RMSE')
axes[1].legend()

plt.tight_layout()
plt.savefig('../data/plot_svd_gridsearch.png', bbox_inches='tight')
plt.show()

print("📌 Key observation:")
print("  RMSE improves up to ~100-150 factors, then plateaus or gets worse.")
print("  This is the bias-variance tradeoff in action:")
print("  Too few factors → underfits (misses real patterns)")
print("  Too many factors → overfits (memorises noise)")
````

````python
# ── Train best SVD ────────────────────────────────────────────────────────────

best_params = gs.best_params['rmse']
print(f"Training SVD with best params: {best_params}")
print()

svd_best = SVD(
    n_factors = best_params['n_factors'],
    n_epochs  = best_params['n_epochs'],
    lr_all    = best_params['lr_all'],
    reg_all   = best_params['reg_all'],
    random_state = 42,
    verbose   = False
)

t0 = time.time()
svd_best.fit(trainset)
train_time = time.time() - t0

predictions = svd_best.test(testset)
rmse_best = accuracy.rmse(predictions, verbose=False)
mae_best  = accuracy.mae(predictions,  verbose=False)

rmse_default_val = accuracy.rmse(svd_default.test(testset), verbose=False)

print(f"Training time     : {train_time:.1f}s")
print(f"RMSE (default)    : {rmse_default_val:.4f}")
print(f"RMSE (tuned)      : {rmse_best:.4f}  (improvement: {rmse_default_val - rmse_best:.4f})")
print(f"MAE  (tuned)      : {mae_best:.4f}")
````

## 6. SVD++ — the improved variant

SVD++ improves on SVD by incorporating **implicit feedback** — the fact that a user rated an item at all (regardless of what score they gave) is itself a signal.

The intuition: if a user rated 500 movies, that alone tells you something about their taste even before looking at the ratings themselves.

**Trade-off:** SVD++ is ~3-5x slower to train than SVD.


````python
print("Training SVD++ (takes longer than SVD)...")
print()

svdpp = SVDpp(
    n_factors = best_params['n_factors'],
    n_epochs  = 20,           # fewer epochs — SVD++ converges more slowly
    lr_all    = best_params['lr_all'],
    reg_all   = best_params['reg_all'],
    random_state = 42,
    verbose   = False
)

t0 = time.time()
svdpp.fit(trainset)
svdpp_time = time.time() - t0

predictions_pp = svdpp.test(testset)
rmse_pp = accuracy.rmse(predictions_pp, verbose=False)
mae_pp  = accuracy.mae(predictions_pp,  verbose=False)

print(f"SVD++  training time : {svdpp_time:.1f}s")
print(f"SVD++  RMSE          : {rmse_pp:.4f}")
print(f"SVD    RMSE          : {rmse_best:.4f}")
print(f"Improvement          : {rmse_best - rmse_pp:.4f}")
print()
print("📌 SVD++ usually gains ~0.01-0.03 RMSE over SVD.")
print("   Whether that trade-off (slower training) is worth it")
print("   depends on your production constraints.")
````

````python
# ── NMF for comparison ────────────────────────────────────────────────────────
print("Training NMF (Non-negative Matrix Factorization)...")
print()

nmf = NMF(
    n_factors = best_params['n_factors'],
    n_epochs  = best_params['n_epochs'],
    random_state = 42,
    verbose   = False
)

t0 = time.time()
nmf.fit(trainset)
nmf_time = time.time() - t0

predictions_nmf = nmf.test(testset)
rmse_nmf = accuracy.rmse(predictions_nmf, verbose=False)

print(f"NMF training time : {nmf_time:.1f}s")
print(f"NMF RMSE          : {rmse_nmf:.4f}")
print()
print("📌 NMF forces all factors to be non-negative.")
print("   This makes factors more interpretable (additive parts-based repr.)")
print("   but usually sacrifices some accuracy vs SVD.")
````

````python
# ── Side by side RMSE comparison ─────────────────────────────────────────────

models_rmse = {
    'SVD (default)': rmse_default_val,
    'SVD (tuned)':   rmse_best,
    'SVD++':         rmse_pp,
    'NMF':           rmse_nmf,
}

fig, ax = plt.subplots(figsize=(8, 4))
colors = [PALETTE[2], PALETTE[0], PALETTE[1], PALETTE[3]]
bars = ax.bar(models_rmse.keys(), models_rmse.values(),
              color=colors, edgecolor='white', width=0.5)
ax.set_title('RMSE comparison across matrix factorization models')
ax.set_ylabel('RMSE (lower is better)')
ax.set_ylim(min(models_rmse.values()) - 0.02, max(models_rmse.values()) + 0.02)
for bar, val in zip(bars, models_rmse.values()):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.002,
            f'{val:.4f}', ha='center', fontsize=10, fontweight='bold')

plt.tight_layout()
plt.savefig('../data/plot_mf_rmse_comparison.png', bbox_inches='tight')
plt.show()
````

## 7. Turn SVD into a `recommend()` function

SVD gives us a predicted rating for any (user, movie) pair. To generate a top-N recommendation list we:
1. Get all movies the user hasn't seen
2. Predict ratings for each one
3. Sort by predicted rating descending
4. Return top-N


````python
def svd_recommend(user_id, n=10, model=None):
    """
    Generate top-N recommendations for a user using a trained SVD model.

    Parameters
    ----------
    user_id : int
    n       : number of recommendations
    model   : trained Surprise model (defaults to svd_best)

    Returns
    -------
    pd.DataFrame with columns: movie_id, title, predicted_rating, source
    """
    if model is None:
        model = svd_best

    # Movies this user has already rated in training
    already_seen = set(train_df[train_df['user_id'] == user_id]['movie_id'])

    # Predict on all unseen movies
    uid_str = str(user_id)
    predictions = []
    for movie_id in all_movie_ids:
        if movie_id in already_seen:
            continue
        iid_str = str(movie_id)
        pred = model.predict(uid_str, iid_str)
        predictions.append((movie_id, pred.est))

    # Sort by predicted rating
    predictions.sort(key=lambda x: x[1], reverse=True)
    top_n = predictions[:n]

    rec_ids    = [p[0] for p in top_n]
    rec_scores = [p[1] for p in top_n]

    recs_df = movies[movies['movie_id'].isin(rec_ids)][['movie_id', 'title']].copy()
    score_map = dict(zip(rec_ids, rec_scores))
    recs_df['predicted_rating'] = recs_df['movie_id'].map(score_map)
    recs_df['source']           = 'SVD'
    recs_df = recs_df.sort_values('predicted_rating', ascending=False).reset_index(drop=True)

    return recs_df


# ── Test ──────────────────────────────────────────────────────────────────────
sample_user = all_user_ids[7]
print(f"SVD recommendations for user {sample_user}:\n")
svd_recs = svd_recommend(sample_user, n=10)
display(svd_recs)

print(f"\nUser {sample_user}'s top-rated training movies:")
user_top = (train_df[train_df['user_id'] == sample_user]
            .nlargest(5, 'rating')
            .merge(movies[['movie_id','title']], on='movie_id')
            [['title','rating']])
display(user_top)
````

````python
# ── Speed: predicting for one user ───────────────────────────────────────────
# Naively iterating over all movies is slow.
# Let's measure and then build a faster version.

print("Timing naive predict-all approach...")
t0 = time.time()
_ = svd_recommend(all_user_ids[0], n=10)
naive_time = time.time() - t0
print(f"  Time for 1 user (naive): {naive_time:.2f}s")
print(f"  Projected for 6040 users: {naive_time * 6040 / 60:.1f} minutes")
print()
print("For production, we pre-compute top-N for all users and cache.")
print("In the API notebook we'll load pre-computed recommendations.")
````

````python
# ── Faster batch version using SVD weight matrices ────────────────────────────
# SVD stores U (user factors) and V (item factors) as numpy arrays.
# We can get ALL predictions for one user with a single matrix multiply.

def svd_recommend_fast(user_id, n=10, model=None):
    """
    Fast SVD recommendation using direct matrix multiplication.
    ~100x faster than the naive predict() loop.
    """
    if model is None:
        model = svd_best

    already_seen = set(train_df[train_df['user_id'] == user_id]['movie_id'])

    # Get inner uid for this user
    uid_str = str(user_id)
    if uid_str not in model.trainset._raw2inner_id_users:
        return popularity_recommend(user_id, n=n)  # unknown user fallback

    inner_uid = model.trainset.to_inner_uid(uid_str)

    # User vector: bias + factors
    u_factors = model.pu[inner_uid]         # shape: (n_factors,)
    u_bias    = model.bu[inner_uid]         # scalar

    # Score all items: global_mean + item_bias + user_bias + u · v
    global_mean = model.trainset.global_mean

    scores = []
    for movie_id in all_movie_ids:
        if movie_id in already_seen:
            continue
        iid_str = str(movie_id)
        if iid_str not in model.trainset._raw2inner_id_items:
            # Unknown item: use global mean + user bias
            score = global_mean + u_bias
        else:
            inner_iid = model.trainset.to_inner_iid(iid_str)
            i_factors = model.qi[inner_iid]
            i_bias    = model.bi[inner_iid]
            score     = global_mean + u_bias + i_bias + np.dot(u_factors, i_factors)
        scores.append((movie_id, float(np.clip(score, 1, 5))))

    scores.sort(key=lambda x: x[1], reverse=True)
    top_n = scores[:n]

    rec_ids    = [p[0] for p in top_n]
    rec_scores = [p[1] for p in top_n]

    recs_df = movies[movies['movie_id'].isin(rec_ids)][['movie_id', 'title']].copy()
    score_map = dict(zip(rec_ids, rec_scores))
    recs_df['predicted_rating'] = recs_df['movie_id'].map(score_map)
    recs_df['source']           = 'SVD'
    recs_df = recs_df.sort_values('predicted_rating', ascending=False).reset_index(drop=True)
    return recs_df


# Verify same results
fast_recs  = svd_recommend_fast(all_user_ids[7], n=10)
naive_recs = svd_recommend(all_user_ids[7], n=10)

print("Fast vs naive — top 5 results match?")
match = all(fast_recs['movie_id'].head(5).values == naive_recs['movie_id'].head(5).values)
print(f"  Match: {match}")
print()

# Time comparison
t0 = time.time()
_ = svd_recommend_fast(all_user_ids[0], n=10)
fast_time = time.time() - t0
print(f"Naive time : {naive_time:.3f}s")
print(f"Fast time  : {fast_time:.4f}s")
print(f"Speedup    : {naive_time/fast_time:.0f}x")
````

## 8. Full evaluation — ranking metrics

````python
print("Evaluating SVD (fast) on 500 users...")
print("(~2-3 minutes)\n")

t0 = time.time()
svd_results = evaluate_model(
    recommend_fn = lambda uid, n: svd_recommend_fast(uid, n),
    user_ids     = all_user_ids,
    k            = 10,
    sample_size  = 500,
)
elapsed = time.time() - t0

# Baseline scores for comparison (from baseline notebook)
pop_results = {
    'precision@k': 0.280,   # replace with your actual values from baseline notebook
    'recall@k':    0.095,
    'ndcg@k':      0.282,
}

print(f"Eval time: {elapsed:.0f}s")
print()
print("=" * 55)
print(f"  {'Model':<22} {'P@10':>8} {'R@10':>8} {'NDCG@10':>10}")
print("-" * 55)
print(f"  {'Popularity baseline':<22} {pop_results['precision@k']:>8.4f} {pop_results['recall@k']:>8.4f} {pop_results['ndcg@k']:>10.4f}")
print(f"  {'SVD (tuned)':<22} {svd_results['precision@k']:>8.4f} {svd_results['recall@k']:>8.4f} {svd_results['ndcg@k']:>10.4f}")
print("=" * 55)
print()

lift = (svd_results['ndcg@k'] - pop_results['ndcg@k']) / pop_results['ndcg@k'] * 100
print(f"NDCG@10 lift over popularity baseline: {lift:+.1f}%")
print()
print("📌 If SVD NDCG ≥ 0.40, we're on track.")
print("   Hybrid model (Week 3) should push this to ≥ 0.50.")
````

## 9. Cold start analysis

Key question: does SVD work equally well for users who rated 5 movies vs users who rated 500 movies?  
This directly informs our alpha blending strategy in the hybrid model.


````python
# ── Segment users by activity level ──────────────────────────────────────────

user_activity = train_df.groupby('user_id')['rating'].count().reset_index()
user_activity.columns = ['user_id', 'n_train_ratings']

bins   = [0, 5, 20, 50, 100, float('inf')]
labels = ['0-5', '6-20', '21-50', '51-100', '100+']
user_activity['segment'] = pd.cut(user_activity['n_train_ratings'],
                                   bins=bins, labels=labels)

print("User segments in training set:")
seg_counts = user_activity['segment'].value_counts()[labels]
for seg, count in seg_counts.items():
    pct = count / len(user_activity) * 100
    print(f"  {seg:<10} : {count:>5,} users  ({pct:.1f}%)")

print()
print("Evaluating SVD per segment (100 users per segment)...")
````

````python
seg_results = {}

for seg in labels:
    seg_users = user_activity[user_activity['segment'] == seg]['user_id'].tolist()
    if len(seg_users) < 10:
        continue

    results = evaluate_model(
        recommend_fn = lambda uid, n: svd_recommend_fast(uid, n),
        user_ids     = seg_users,
        k            = 10,
        sample_size  = min(100, len(seg_users)),
        random_state = 42,
    )
    seg_results[seg] = results
    print(f"  {seg:<10} : NDCG@10 = {results['ndcg@k']:.4f}  (n={results['n_users']})")

print()

# ── Visualise ─────────────────────────────────────────────────────────────────
fig, axes = plt.subplots(1, 2, figsize=(13, 4))

segs_plot  = [s for s in labels if s in seg_results]
ndcg_vals  = [seg_results[s]['ndcg@k']      for s in segs_plot]
prec_vals  = [seg_results[s]['precision@k'] for s in segs_plot]

axes[0].bar(segs_plot, ndcg_vals, color=PALETTE[0], edgecolor='white')
axes[0].set_title('SVD NDCG@10 by user activity segment')
axes[0].set_xlabel('Training ratings per user')
axes[0].set_ylabel('NDCG@10')
axes[0].axhline(y=pop_results['ndcg@k'], color=PALETTE[3],
                linestyle='--', linewidth=1.5, label='Popularity baseline')
axes[0].legend()
for i, (seg, val) in enumerate(zip(segs_plot, ndcg_vals)):
    axes[0].text(i, val + 0.003, f'{val:.3f}', ha='center', fontsize=9)

axes[1].bar(segs_plot, prec_vals, color=PALETTE[1], edgecolor='white')
axes[1].set_title('SVD Precision@10 by user activity segment')
axes[1].set_xlabel('Training ratings per user')
axes[1].set_ylabel('Precision@10')
for i, (seg, val) in enumerate(zip(segs_plot, prec_vals)):
    axes[1].text(i, val + 0.001, f'{val:.3f}', ha='center', fontsize=9)

plt.tight_layout()
plt.savefig('../data/plot_svd_cold_start.png', bbox_inches='tight')
plt.show()

print("📌 KEY FINDING:")
print("  SVD performance drops sharply for users with < 20 ratings.")
print("  For 0-5 rating users, SVD barely beats the popularity baseline.")
print()
print("  → This is exactly WHY we need the hybrid:")
print("    low-activity users: lean on content-based (α = 0.2)")
print("    high-activity users: lean on SVD (α = 0.8)")
print()
print("  This cold start chart is one of the strongest results")
print("  to show in interviews — it proves you thought about production.")
````

## 10. Explore the latent factors

Let's peek inside what SVD actually learned.

````python
# ── What do the latent factors look like? ────────────────────────────────────
# We can project movie factors onto 2D using PCA and see if genres cluster.

from sklearn.decomposition import PCA

# Get movie factor matrix from trained SVD
# trainset maps raw movie_id (str) → inner_id
inner_ids    = []
movie_ids_in = []

for movie_id in all_movie_ids:
    iid_str = str(movie_id)
    if iid_str in svd_best.trainset._raw2inner_id_items:
        inner_ids.append(svd_best.trainset.to_inner_iid(iid_str))
        movie_ids_in.append(movie_id)

V_matrix = svd_best.qi[inner_ids]  # shape: (n_movies_seen, n_factors)
print(f"Movie factor matrix shape: {V_matrix.shape}")

# PCA to 2D for visualisation
pca = PCA(n_components=2, random_state=42)
V_2d = pca.fit_transform(V_matrix)
print(f"Explained variance: PC1={pca.explained_variance_ratio_[0]:.2%}, PC2={pca.explained_variance_ratio_[1]:.2%}")
````

````python
# ── Colour by genre ───────────────────────────────────────────────────────────
movie_factors_df = pd.DataFrame({
    'movie_id': movie_ids_in,
    'pc1':      V_2d[:, 0],
    'pc2':      V_2d[:, 1],
}).merge(movies[['movie_id', 'title', 'genres']], on='movie_id')

# Assign a primary genre to each movie
def primary_genre(genres_str):
    g = genres_str.split('|')[0]
    return g

movie_factors_df['primary_genre'] = movie_factors_df['genres'].apply(primary_genre)

# Plot top genres only
top_genres = movie_factors_df['primary_genre'].value_counts().head(8).index.tolist()
plot_df = movie_factors_df[movie_factors_df['primary_genre'].isin(top_genres)]

fig, ax = plt.subplots(figsize=(10, 7))
for i, genre in enumerate(top_genres):
    sub = plot_df[plot_df['primary_genre'] == genre]
    ax.scatter(sub['pc1'], sub['pc2'], label=genre,
               alpha=0.5, s=20, color=PALETTE[i % len(PALETTE)])

ax.set_title('SVD movie latent factors projected to 2D (PCA)')
ax.set_xlabel(f'PC1 ({pca.explained_variance_ratio_[0]:.1%} variance)')
ax.set_ylabel(f'PC2 ({pca.explained_variance_ratio_[1]:.1%} variance)')
ax.legend(markerscale=2, fontsize=9)

plt.tight_layout()
plt.savefig('../data/plot_svd_latent_space.png', bbox_inches='tight')
plt.show()

print("📌 If genres cluster in the 2D PCA space,")
print("   it means SVD's latent factors captured genre-like structure")
print("   purely from rating patterns — without ever seeing genre labels.")
````
